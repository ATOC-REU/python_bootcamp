## 1) Import the numpy and matplotlib packages for this challenge

```python
import numpy as np
import matplotlib.pyplot as plt
```

Here we read in the data. Don't worry yet about how to do this (you'll learn later!).

After we do this, we have 4 different arrays: <br>
>1) temp = array of daily mean air temperature values (degrees Fahrenheit)<br>
>2) years = array of years for each temperature value <br>
>3) months = array of months for each temperature value <br>
>4) days = array of days for each temperature value <br>

```python
import pandas as pd
d = pd.read_csv('boulderdaily.complete.txt', delim_whitespace=True, skiprows = 1, usecols = [0,1,2,3,4], names = ['day#', 'year', 'month', 'day', 'T'])

temp = d['T'].values
years = d.year.values
months = d.month.values
days = d.day.values
```

What is the first and last day, month, year in this data set?

```python
print('First day:', days[0], months[0], years[0])
print('Last day:', days[-1], months[-1], years[-1])
```

## 2) Convert the data from Fahrenheit to Celsius and plot

Create a new array by converting the temperature values to Celsius using: $C = \frac{5}{9}*(F-32)$

```python
temp_c = (5/9)*(temp-32)
```

Plot the Celsius temperature data. Label the x-axis "day #" and the y-axis "Temp (C)"

```python
fig, ax = plt.subplots(figsize = (15,5))
ax.plot(temp)
ax.set_xlabel('day #', fontsize = 16)
ax.set_ylabel('Temp (C)', fontsize = 16)
```

## 3) What is the minimum, maximum, mean, and standard deviation of this celsius temperature dataset?


```python
print('Minimum: ', np.min(temp_c))
print('Maximum: ', np.max(temp_c))
print('Mean: ', np.mean(temp_c))
print('Std: ', np.std(temp_c))
```

## 4) Select the temperature values from the year 2018 and 2019

Create a new array of all temperature values from 2018, and another one for 2019

Hint: see https://www.w3schools.com/python/numpy/numpy_array_filter.asp

```python
temp2018 = temp_c[years == 2018]
temp2019 = temp_c[years == 2019]
```

Plot both years on the same plot. Label the x and y-axis the same as the plot above. Label the two lines and add a legend.

```python
fig, ax = plt.subplots(figsize = (15,5))
ax.plot(temp2018, label = '2018')
ax.plot(temp2019, label = '2019')
ax.set_xlabel('day #', fontsize = 16)
ax.set_ylabel('Temp (F)', fontsize = 16)
ax.legend()
```

Which year had a higher mean temperature?

In which year was temperature more variable?

```python
print(np.mean(temp2018))
print(np.mean(temp2019))
print('2018 had a higher mean temperature')
```

```python
print(np.std(temp2018))
print(np.std(temp2019))
print('2019 had a more variable temperature')
```

## 5) Comparing seasonal data

Create an array of 2019 temperatures for just the winter months (Jan, Feb) and another array of just the summer months (June, July)

```python
months2019 = months[years==2019]
winter2019 = temp2019[(months2019==1)|(months2019==2)]
summer2019 = temp2019[(months2019==6)|(months2019==7)]
```

Plot both seasons on the same plot. Label the two lines and add a legend.

```python
plt.figure(figsize = (15,5))
plt.plot(winter2019, label = 'winter')
plt.plot(summer2019, label = 'summer')
plt.xlabel('day #', fontsize = 16)
plt.ylabel('Temp (F)', fontsize = 16)
plt.legend()
```

What is the mean temperature for each season?

Which season has a more variable temperature?

```python
print('Summer average temp = ', summer2019.mean())
print('Winter average temp = ', winter2019.mean())

print(summer2019.std())
print(winter2019.std())
print('Winter is more variable')
```

## 6) Plotting challenge!

Try to recreate the plot below as closely as possible:



```python
x = np.linspace(-10,10)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x**2
y4 = x**3

fig, ax = plt.subplots(1,2, figsize = (15,5))
ax[0].plot(x,y1, color = 'skyblue', linewidth = 4, label = 'sin(x)')
ax[0].plot(x,y2, color = 'darkblue', linestyle = '--', label = 'cos(x)')
ax[0].legend()
ax[0].set_ylabel('Y', fontsize = 16)
ax[0].set_xlabel('X', fontsize = 16)
ax[0].set_xticks([-10,-5,0,5,10])
ax[0].set_yticks([-1,-0.5,0,0.5,1])
ax[0].set_title('My plot #1', fontsize = 20)

ax[1].plot(x,y3, color = 'maroon', linewidth = 4, label = 'x**2')
ax[1].plot(x,y4, color = 'lightcoral', linestyle = '--', label = 'x**3')
ax[1].legend(fontsize = 14)
ax[1].set_ylabel('Y', fontsize = 16)
ax[1].set_xlabel('X', fontsize = 16)
ax[1].set_xticks([-10,-5,0,5,10])
ax[1].grid()
ax[1].set_title('My plot #2', fontsize = 20)

plt.savefig('myfig.png')
```
